#Rodando o Dummy Classifier

```python
x = dados_binario.drop("Churn",axis=1)
y = dados_binario["Churn"]
```

```python
#Dummy classifier
from sklearn.model_selection import train_test_split
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score
from imblearn.over_sampling import SMOTE
from sklearn.metrics import classification_report, confusion_matrix

SEED = 20

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y,
                                                         random_state = SEED,
                                                         stratify = y)

#Smote
smote = SMOTE(sampling_strategy=0.5 , random_state=20)
x_treino , y_treino = smote.fit_resample(x_treino,y_treino)


print(f"Treinaremos com {len(x_treino)}")
print(f"Testaremos com {len(x_teste)}")

classificador = DummyClassifier()
classificador.fit(x_treino,y_treino)
previsoes = classificador.predict(x_teste)

acuracia = accuracy_score(y_teste, previsoes) * 100
print(f"A acurácia do dummy foi de {acuracia:.2f}%")



print("\nMatriz de confusão:")
print(confusion_matrix(y_teste, previsoes))

print("\nRelatório de classificação:")
print(classification_report(y_teste, previsoes, zero_division=0))
```

Sem fazer o balanceamento dos dados é possível perceber que o modelo utilizando o dummy classifier (um classificador bobo) acertaria 73% dos casos, já que existe mais dados 0 (não) do 1 (sim)

```python
def acuracia_SVC(y_teste,previsoes):
  from sklearn.metrics import accuracy_score
  return accuracy_score(y_teste,previsoes)
```

```python
def precisao_SVC(y_teste,previsoes):
  from sklearn.metrics import precision_score
  return precision_score(y_teste,previsoes)
```

```python
def recall_SVC(y_teste,previsoes):
  from sklearn.metrics import recall_score
  return recall_score(y_teste,previsoes)
```

```python
def matriz_confusao_SVC(y_teste,previsoes):
  from sklearn.metrics import confusion_matrix
  return confusion_matrix(y_teste,previsoes)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC


SEED = 20
x_treino, x_teste , y_treino , y_teste = train_test_split(x , y , test_size = 0.3 , random_state = 20)


smote = SMOTE(sampling_strategy=0.5 , random_state=20)
x_treino , y_treino = smote.fit_resample(x_treino,y_treino)

# print(f"Treinaremos com {len(x_treino)}")
# print(f"Testaremos com {len(x_teste)}")


scaler = StandardScaler()
scaler.fit(x_treino)

x_treino = scaler.transform(x_treino)
x_teste = scaler.transform(x_teste)


modelo = LinearSVC()
modelo.fit(x_treino, y_treino)
previsoes = modelo.predict(x_teste)

#SVC
print(f"Acurácia SVC: {acuracia_SVC(y_teste,previsoes)}")
print(f"Recall SVC: {recall_SVC(y_teste,previsoes)}")
print(f"Precisão da SVC: {precisao_SVC(y_teste,previsoes)}")
print(f"Matriz de confusão SVC : \n{matriz_confusao_SVC(y_teste,previsoes)}")
print("="*30)
print("\nRelatório de classificação:")
print(classification_report(y_teste, previsoes, zero_division=0))
```

```python
def exec_classificador(classificador_arvore_decisao, x_treino,x_teste,y_treino):
  from sklearn.tree import DecisionTreeClassifier
  arvore =classificador_arvore_decisao.fit(x_treino,y_treino)
  y_pred = arvore.predict(x_teste)
  return y_pred
```

```python
def acuracia(y_teste,y_pred_arvore):
  from sklearn.metrics import accuracy_score
  return accuracy_score(y_teste,y_pred_arvore)
```

```python
def matriz_confusao(y_teste,y_pred_arvore):
  from sklearn.metrics import confusion_matrix
  return confusion_matrix(y_teste,y_pred_arvore)
```

```python
def precisao_arvore(y_teste,y_pred_arvore):
  from sklearn.metrics import precision_score
  return precision_score(y_teste,y_pred_arvore)
```

```python
def recall_arvore(y_teste,y_pred_arvore):
  from sklearn.metrics import recall_score
  return recall_score(y_teste,y_pred_arvore)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
classificador_arvore_decisao = tree.DecisionTreeClassifier(max_depth=10, random_state=20,min_samples_leaf=5,class_weight="balanced")
y_pred_arvore =  exec_classificador(classificador_arvore_decisao , x_treino, x_teste, y_treino)
```

```python
#Arvore de decisão
print(f"Acurácia: {acuracia(y_teste,y_pred_arvore)}")
print(f"Recall árvore: {recall_arvore(y_teste,y_pred_arvore)}")
print(f"Precisão da árvore: {precisao_arvore(y_teste,y_pred_arvore)}")
print(f"Matriz de confusão : \n{matriz_confusao(y_teste,y_pred_arvore)}")
```

#Hiper parâmetros de duas dimensões

```python

def roda_arvore(max_depth,min_samples_leaf):
  from sklearn.model_selection import KFold
  from sklearn.tree import DecisionTreeClassifier
  from sklearn.model_selection import cross_validate
  import numpy as np

  SEED = 301
  np.random.seed(SEED)

  cv = KFold(n_splits = 10)
  modelo = DecisionTreeClassifier(max_depth=max_depth,min_samples_leaf=min_samples_leaf)
  results = cross_validate(modelo, x, y, cv = cv, return_train_score=True)

  #Variaveis
  test_score = round(results['test_score'].mean() * 100,2)
  train_score = round(results['train_score'].mean() * 100,2)

  #resultados
  #print("="*30)
  #print(f"Arvore max_depth={max_depth} \n min_samples_leaf: {min_samples_leaf} \n Treino Média: {train_score} \n Teste média: {test_score}")
  #print("="*30)
  tabela = [max_depth, min_samples_leaf, train_score , test_score ]

  return tabela


def busca():
  resultados = []
  for max_depth in range(1,33):
    for min_samples_leaf in [32,64,128,256]:
      tabela = roda_arvore(max_depth,min_samples_leaf)
      resultados.append(tabela)
  resultados = pd.DataFrame(data=resultados , columns=['max_depth', 'min_samples_leaf', 'train_score' , 'test_score'])
  return resultados

```

#Testando a terceira dimensão de hiper parâmetros

```python

def roda_arvore(max_depth,min_samples_leaf,min_samples_split):
  from sklearn.model_selection import KFold
  from sklearn.tree import DecisionTreeClassifier
  from sklearn.model_selection import cross_validate
  import numpy as np

  SEED = 301
  np.random.seed(SEED)

  cv = KFold(n_splits = 10)
  modelo = DecisionTreeClassifier(max_depth=max_depth,min_samples_leaf=min_samples_leaf,min_samples_split=min_samples_split)
  results = cross_validate(modelo, x, y, cv = cv, return_train_score=True)
  fit_time = results["fit_time"].mean()
  score_time = results["score_time"].mean()
  #Variaveis
  test_score = round(results['test_score'].mean() * 100,2)
  train_score = round(results['train_score'].mean() * 100,2)

  #resultados
  tabela = [max_depth, min_samples_leaf,min_samples_split, train_score ,
            test_score ,fit_time, score_time]

  return tabela


def busca():
  resultados = []
  for max_depth in range(1,33):
    for min_samples_leaf in [128,192,256,512]:
      for min_samples_split in [32, 64 ,128 ,256]:
        tabela = roda_arvore(max_depth,min_samples_leaf,min_samples_split)
        resultados.append(tabela)
  resultados = pd.DataFrame(data=resultados , columns=['max_depth', 'min_samples_leaf', 'min_samples_split','train_score' , 'test_score','fit_time','score_time'])
  return resultados
```
